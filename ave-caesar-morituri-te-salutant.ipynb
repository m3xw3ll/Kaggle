# <span style='color:#f6416c; font-size: 40px; font-weight: bold; line-height: 40px;'>|</span> <span style='color:#222831; font-size: 40px; font-weight: bold;'>Introduction</span>
Gladiators, the iconic combatants of ancient Rome, were warriors who fought in brutal arenas for the entertainment of the public. These fighters came from diverse backgrounds—some were prisoners of war, others slaves, or even volunteers seeking fame and fortune. While gladiatorial combat was often a fight to the death, many gladiators managed to survive numerous bouts, earning their freedom or a revered place in history.

In this project, we explore a rich dataset of ancient gladiators, detailing various aspects of their lives, physical attributes, fighting styles, and social status. From their origin and combat experience to their patron’s wealth and public favor, each gladiator has a unique set of characteristics that influenced their fate in the arena. Our goal is to analyze these attributes to predict a critical outcome—whether a gladiator survived the harsh and unforgiving gladiatorial games.

By leveraging data such as age, weight, weapon choice, mental resilience, and even crowd appeal techniques, we aim to uncover the factors that most significantly impacted a gladiator's chances of survival. The predictions derived from this analysis will help us understand the complex interplay of physical prowess, strategy, and social dynamics that dictated life or death for these ancient fighters.

## <span style='color:#f6416c; font-size: 30px; font-weight: bold; line-height: 30px;'>|</span> <span style='color:#222831; font-size: 30px; font-weight: bold;'>Chi-Square Selection</span>

The Chi-Square test is a statistical method commonly used in feature selection to evaluate the association between categorical features and the target variable. The main goal of this test is to determine whether the presence or absence of a certain feature has a significant relationship with the outcome.

In the context of machine learning, particularly classification problems, feature selection is crucial to ensure that your model is built on the most relevant information. The Chi-Square test helps to identify which categorical features are most strongly associated with the target variable.

Here's how it works: The test compares the observed frequencies of each feature-target combination (what is present in your dataset) with the expected frequencies (what would be expected if the feature and target were independent). Essentially, it checks whether the distribution of the target variable changes when different values of the feature are observed.

If the observed frequencies are very different from the expected ones, it suggests that the feature might have a significant impact on the target variable. In this case, the Chi-Square statistic (a numerical value) will be high, and the p-value (which indicates the likelihood that the observed difference is due to chance) will be low. A low p-value (typically less than 0.05) suggests that the feature is likely relevant for predicting the target, as there is a significant association between the two.

The Chi-Square test is particularly effective for categorical features because it operates by comparing frequencies, making it suitable for data where features represent categories like names, locations, or types. If the test finds that the presence of a certain feature category (e.g., "Gaul" as the origin of a gladiator) makes a difference in whether or not the gladiator survives, that feature is deemed important.

In practice, after running the Chi-Square test for each feature in your dataset, you can rank the features based on their Chi-Square statistic or p-value. Features that have a high Chi-Square statistic and a low p-value are considered to have a strong association with the target variable and are likely useful for making predictions.

However, the test has some limitations. It can only be applied to categorical features, and all features must be encoded numerically before performing the test. Additionally, it assumes that the observations in your dataset are independent of each other, which might not always be the case. While the Chi-Square test is a valuable tool, it doesn’t consider interactions between features, so it’s often used alongside other feature selection methods to get a comprehensive view of feature importance.

# <span style='color:#f6416c; font-size: 40px; font-weight: bold; line-height: 40px;'>|</span> <span style='color:#222831; font-size: 40px; font-weight: bold;'>Cross Validation</span>

```python
def cross_validation_score(estimator, cv, label=''):
    X = train.copy()
    y = encoder.transform(X.pop('Survived'))
    X = pd.get_dummies(X)
    
    train_scores, val_scores = [], []
    for fold, (train_idx, val_idx) in enumerate(cv.split(X, y)):
        model = clone(estimator)
        X_train = X.iloc[train_idx].reset_index(drop=True)
        y_train = y[train_idx]
        X_val = X.iloc[val_idx].reset_index(drop=True)
        y_val = y[val_idx]
        
        model.fit(X_train, y_train)

        train_preds = model.predict_proba(X_train)[:, 1]
        val_preds = model.predict_proba(X_val)[:, 1]
                
        train_score = roc_auc_score(y_train, train_preds)
        val_score = roc_auc_score(y_val, val_preds)
        train_scores.append(train_score)
        val_scores.append(val_score)
    print(f'Val Score: {np.mean(val_scores):.5f} ± {np.std(val_scores):.5f} | Train Score: {np.mean(train_scores):.5f} ± {np.std(train_scores):.5f} | {label}')
    return val_scores
```

<div style="border-radius:10px; border:#ff9234 solid; padding: 15px; background-color: #f8f3d4; font-size:100%; text-align:left">

<h3 align="left"><font color='#d92027'>💬 Conclusion</font></h3>

Cross validation is an important step in machine learning because it provides an unbiased evaluation of a model´s performance by splitting the dataset into multiple training and validation sets (or folds). This process helps detect overfitting, where the model performs well on the training data but poorly on unseen data. By training and validating the model on different subsets of the data, cross-validation ensures that the evaluation metrics are representative of the model's generalization ability. Additionally, cross-validation allows for hyperparameter tuning and model comparison, helping select the best-performing model.

```python
score_list, oof_list = pd.DataFrame(), pd.DataFrame()

models = [
    ('xgb', XGBClassifier(random_state = seed)),
    ('lgb', LGBMClassifier(random_state = seed, verbose=-1)),
    ('cb', CatBoostClassifier(random_state = seed, verbose = 0)),
    ('gb', GradientBoostingClassifier(random_state = seed)),
    ('hgb', HistGradientBoostingClassifier(random_state = seed)),
]
```

```python
for (label, model) in models:
    score_list[label] = cross_validation_score(estimator=model,
                                                                cv=skf,
                                                                label=label)
```

The **Voting Classifier** is an ensemble method that combines predictions from multiple models to improve overall performance and robustness. It works by taking the majority vote (hard voting) or averaging class probabilities (soft voting) to make the final prediction. This technique is valuable because it leverages the strengths of each individual model while minimizing their weaknesses, making the ensemble more reliable and accurate than any single model. By combining diverse models, the Voting Classifier reduces the risk of overfitting, enhances generalization, and provides better predictive performance, especially when base models perform well on different parts of the data.

```python
cb_model = CatBoostClassifier(random_state = seed, verbose = 0)
lgb_model = LGBMClassifier(random_state = seed, verbose=-1)
hgb_model = HistGradientBoostingClassifier(random_state = seed)
voting = VotingClassifier(estimators=[
    ('cb', cb_model),
    ('lgb', lgb_model),
    ('hgb', hgb_model)
], voting='soft')
```

# <span style='color:#f6416c; font-size: 40px; font-weight: bold; line-height: 40px;'>|</span> <span style='color:#222831; font-size: 40px; font-weight: bold;'>Threshold Moving</span>

Threshold moving is a technique used in binary classification to adjust the decision boundary by changing the probability threshold that determines whether a sample is classified as positive or negative. In most classifiers, the default threshold is set at 0.5, meaning that samples with a predicted probability greater than or equal to 0.5 are labeled as positive, while those below 0.5 are labeled as negative. However, in many real-world scenarios, such as when dealing with imbalanced datasets or cases where the cost of false positives and false negatives is not equal, this default threshold may not yield the best performance. For example, in medical diagnosis, it might be more important to reduce false negatives (i.e., failing to identify a disease) than false positives, necessitating a lower threshold to increase recall at the expense of precision.

To find the optimal threshold, one can experiment by varying the threshold value and observing how it impacts different performance metrics such as precision, recall, F1 score, or ROC-AUC. The goal is to identify the threshold that maximizes a specific metric or achieves the best balance between conflicting metrics like precision and recall. One effective approach is to calculate the F1 score at various thresholds and choose the threshold that maximizes this score. The ROC curve, which plots the true positive rate (sensitivity) against the false positive rate, can also be used to visualize the model’s performance at different thresholds. The best threshold can then be determined by maximizing the F1 score or by using Youden’s J statistic, which represents the point on the ROC curve that is farthest from the diagonal line of random classification.

```python
preds_proba = preds_proba[:, 1]
thresholds = np.linspace(0, 1, 100)
f1_scores = [f1_score(y_test, preds_proba >= t) for t in thresholds]

best_threshold = thresholds[np.argmax(f1_scores)]
```

```python
plt.figure(figsize=(8, 6))
plt.plot(thresholds, f1_scores, label='F1 Score', color='#222831')
plt.scatter(best_threshold, max(f1_scores), color='#f6416c', label=f'Best Threshold = {best_threshold:.2f}')
plt.xlabel('Threshold')
plt.ylabel('F1 Score')
plt.title('F1 Score vs Threshold')
plt.legend()
plt.show()
```

```python
fpr, tpr, thresholds = roc_curve(y_test, preds_proba)
gmeans = sqrt(tpr * (1-fpr))
ix = argmax(gmeans)
```

```python
plt.plot(fpr, tpr, color='#222831')
plt.scatter(fpr[ix], tpr[ix], color='#f6416c', label=f'Best Threshold = {thresholds[ix]:.2f}')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

<h3 align="left"><font color='#d92027'>💬 Conclusion</font></h3>

Threshold moving is used to modify the decision boundary of a binary classifier. If the dataset is highly imbalanced a normal threshold (0.5) might not be suitable as it can lead to high precision but poor recall, or vica versa. We have several methods to find the best threshold. I used two different approaches to calculate the threshold:
    
- **Maximizing a Metric:** Finding the threshold that maximizes a metric like F1 score or precision-recall. (threshold at 0.37)
- **ROC Curve Analysis:** Choosing a threshold based on the ROC curve and Youden's J statistic, which is sensitivity + specificity - 1. (threshold at 0.45)

## <span style='color:#f6416c; font-size: 30px; font-weight: bold; line-height: 30px;'>|</span> <span style='color:#222831; font-size: 30px; font-weight: bold;'>Classification Report</span>

The classification report is a comprehensive summary that provides several key performance metrics for evaluating a classification model. It is typically used for binary and multi-class classification problems. The report is divided into several rows and columns, with each row representing a class label (e.g., 0 and 1 in a binary classification) and each column containing specific metrics like precision, recall, F1-score, and support.

The metrics presented in a classification report help understand how well the classifier performed for each class and overall. 

<div style="border-radius:10px; border:#ff9234 solid; padding: 15px; background-color: #f8f3d4; font-size:100%; text-align:left">

<h3 align="left"><font color='#d92027'>💬 Components</font></h3>
    
1. <span style="color: #d92027; font-weight: bold;">Precision:</span>
    - Precision indicates the percentage of correctly predicted positive instances out of all instances predicted as positive
    - Precision is crucial when the cost of false positives is high, such as in spam detection, where you want to avoid incorrectly classifying legitimate emails as spam
    
2. <span style="color: #d92027; font-weight: bold;">Recall:</span>
    - Recall (also known as Sensitivity or True Positive Rate) measures the percentage of correctly predicted positive instances out of all actual positive instances
    - Recall is essential when the cost of false negatives is high, such as in medical diagnostics, where missing a positive case (e.g., cancer detection) can have severe consequences
    
3. <span style="color: #d92027; font-weight: bold;">F1-Score:</span>
    - The F1-Score is the harmonic mean of precision and recall. It provides a balanced metric that considers both false positives and false negatives, making it useful when the class distribution is imbalanced
    
4. <span style="color: #d92027; font-weight: bold;">Support:</span>
    - Support indicates the number of true instances for each class in the dataset
    - It is simply the count of occurrences of each class in y_true
    - While support is not a performance metric, it is helpful for understanding the distribution of class labels
    
5. <span style="color: #d92027; font-weight: bold;">Macro Average:</span>
    - The macro average computes the metric independently for each class and then takes the average, treating all classes equally
    - It is useful when you want to assess how well the model performs across all classes without considering class imbalance
    
6. <span style="color: #d92027; font-weight: bold;">Weighted Average:</span>
    - The weighted average takes the metrics for each class and computes a weighted mean, with weights corresponding to the number of true instances for each class (i.e., support)
    - This is beneficial when dealing with class imbalance, as it gives more importance to classes with higher support
